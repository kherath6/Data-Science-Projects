```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

```python
DATA_DIR = '../healthcare'
FILE_NAME = 'healthcareClanedTrain3.csv'
data_path = os.path.join(DATA_DIR, FILE_NAME)
health = pd.read_csv(data_path)
```

```python
healthFeats = health.drop('stay', axis = 1)
```

```python
healthFeats.head()
```

```python
# using scikit-learn to perorm k-means clustering
from sklearn.cluster import KMeans

# specify the number of clusters (2) adn fit the data dat_rider_feats
kmeans = KMeans(n_clusters = 2, random_state = 0).fit(healthFeats)
```

```python
centroids = (kmeans.cluster_centers_)
print(centroids)
```

```python
labels = (kmeans.labels_)
print(labels)
```

```python
# calculate silhouette_score
#from sklearn.metrics import silhouette_score
#print(silhouette_score(healthFeats, kmeans.labels_))
```

```python
df_labels = pd.DataFrame(labels)
df_labels.rename(columns={0:'Cluster'}, inplace = True)
```

```python
healthClust = pd.concat([health, df_labels], axis = 1)
```

```python
healthClust.head()
```

```python
healthClustC0 = healthClust.loc[healthClust['Cluster'] == 0]
healthClustC0.head()
```

```python
healthClustC1 = healthClust.loc[healthClust['Cluster'] == 1]
healthClustC1.head()
```

```python
healthC0 = healthClustC1.drop('Cluster', axis = 1)
healthC0.head()
```

```python
healthC1 = healthClustC1.drop('Cluster', axis = 1)
healthC1.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
class_tree = DecisionTreeClassifier(max_depth = 3)
class_tree.fit(health.drop('stay', axis = 1), health['stay'])
```

```python
from sklearn.metrics import confusion_matrix
y_train_pred = class_tree.predict(health.drop('stay', axis = 1))
confusion_matrix(health['stay'], y_train_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true = health['stay'], y_pred = y_train_pred)
accuracy.round(3)
```

```python
class_tree_C0 = DecisionTreeClassifier(max_depth = 3)
class_tree_C0.fit(healthC0.drop('stay', axis = 1), healthC0['stay'])
y_train_pred_C0 = class_tree.predict(healthC0.drop('stay', axis = 1))
accuracy = accuracy_score(y_true = healthC0['stay'], y_pred = y_train_pred_C0)
accuracy.round(3)
```

```python
class_tree_C1 = DecisionTreeClassifier(max_depth = 3)
class_tree_C1.fit(healthC1.drop('stay', axis = 1), healthC1['stay'])
y_train_pred_C1 = class_tree.predict(healthC1.drop('stay', axis = 1))
accuracy = accuracy_score(y_true = healthC1['stay'], y_pred = y_train_pred_C1)
accuracy.round(3)
```
